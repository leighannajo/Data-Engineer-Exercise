```python
# import pandas
import pandas as pd

# import and ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

Read in the constituent email address file.

```python
df_emails = pd.read_csv('data/cons_email.csv')
print(df_emails.shape)
df_emails.head()
```

Filter dataset to primary email addresses only.

```python
df_emails = df_emails.loc[df_emails['is_primary']==1]
df_emails
```

Check the data types of the email address dataframe.

```python
df_emails.dtypes
```

Check the column names of the email address dataset.

```python
df_emails.columns
```

Read in the subscription status file and filter to statuses where chapter_id is 1 only.

```python
df_sub = pd.read_csv('data/cons_email_chapter_subscription.csv')
df_sub = df_sub.loc[df_sub['chapter_id']==1]
df_sub
```

Check the dataframe types.

```python
df_sub.dtypes
```

Sanity check --> test unique chapter_id statuses to ensure only status of 1 is present.

```python
df_sub.chapter_id.unique()
```

Check the column names of the subsription dataset.

```python
df_sub.columns
```

Merge the email address and subscription datasets on the common column cons_email_id.

```python
df_subEmail = df_emails.merge(df_sub, how='inner', 
                            on='cons_email_id')
df_subEmail
```

Check the current datatypes of the new merged dataset.

```python
df_subEmail.dtypes
```

```python
df_subEmail.cons_id.unique()
```

Change datatype of the cons_id column by replacing null values with 0 and converting to integers.

```python
df_subEmail.cons_id.fillna(0, inplace=True)
```

```python
df_subEmail.cons_id = df_subEmail.cons_id.astype('int')
```

```python
df_subEmail.dtypes
```

Read in the constituents dataset.

```python
df_cons = pd.read_csv('data/cons.csv')
df_cons.head()
```

```python
df_cons.columns
```

```python
df_cons.source.unique()
```

Merge the new matched dataset with the constitiuents dataset on the cons_id column.

```python
df_all = df_subEmail.merge(df_cons, how='inner', 
                            on='cons_id')
df_all
```

```python
df_all.columns
```

Create an empty dataframe for the final people file using the needed columns.

```python
df_final = pd.DataFrame()
```

```python
df_final[['email', 'code', 'is_unsub', 
          'created_dt', 'updated_dt']] = df_all[['email', 'source', 
                                                    'isunsub', 'create_dt_y', 'create_dt_x']]
```

```python
df_final
```

Check the datatypes of the newly created people dataset.

```python
df_final.dtypes
```

Convert the is_unsub column to boolean datatype and the created_dt and updated_dt columns to datetime objects.

```python
df_final['is_unsub'] = df_final['is_unsub'].astype('bool')
df_final['created_dt']= pd.to_datetime(df_final['created_dt'])
df_final['updated_dt']= pd.to_datetime(df_final['updated_dt'])
```

```python
df_final.dtypes
```

Create the acquisitions file by grouping on date of the created_dt column and counting all instances of constituents acquired on that date.

```python
df_acquistions = df_final.groupby(pd.Grouper(key='created_dt', axis=0, 
                      freq='2D', sort=True)).count()
df_acquistions
```

Create the final acquistion file by resetting the index of the grouped data, creating an empty dateframe and then creating the columns needed.

```python
df_acquistions.reset_index(inplace=True)
```

```python
df_acq_final = pd.DataFrame()
```

```python
df_acq_final[['acquisition_date', 
              'acquistions']] = df_acquistions[['created_dt', 'email']]
```

```python
df_acq_final
```

Save both dataframes as csv files.
